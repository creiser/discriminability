This notebook creates subsets of ImageNet.
ImageNet must be placed in **datasets/imagenet-full** before this notebook can be run.

```python
hp = {}
hp['num_datasets'] = 10
hp['num_categories_per_dataset'] = 10
```

```python
import os
from twosample import utils

# Filter all categories that do not contain 1300 images
imagenet_directory = 'datasets/imagenet-full'
candidate_categories = []
for category in os.listdir(imagenet_directory):
    num_images = len(os.listdir(imagenet_directory + '/' + category))
    if num_images == 1300:
        candidate_categories.append(category)
    print(category, num_images)
print('#candidate categories:', len(candidate_categories))
```

```python
# Construct datasets by selecting 10 categories at random
for dataset_index in range(hp['num_datasets']):
    dataset_categories = utils.random_subset(candidate_categories, hp['num_categories_per_dataset'])
    print(dataset_categories)
    
    dataset_directory = 'datasets/imagenet-{}'.format(dataset_index)
    utils.create_directory_if_not_exists(dataset_directory)
    
    for category in dataset_categories:
        source_directory = '../imagenet-full/' + category
        destination_directory = dataset_directory + '/' + category
        print(source_directory, destination_directory)
        os.symlink(source_directory, destination_directory)
```
